```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib qt

df = pd.read_csv('weatherHistory_univariate.csv')
df.head(20)
```

```python
df = df.sort_values('Formatted Date')
df.head(30)
```

```python
# Converting the dt column to datetime object 
df['Formatted Date'] = [pd.to_datetime(x) for x in df['Formatted Date']]
df
```

```python
df.plot(x='Formatted Date', y='Temperature (C)', kind='line',
        figsize=(8,4),
        title='Temperature (C)')
```

```python
time1 = pd.to_datetime('2006-01-01 00:00:00+01:00', format='%Y-%m-%d', utc=True)
time1
```

```python
time2 = pd.to_datetime('2006-01-31 23:59:59+00:00', format='%Y-%m-%d', utc=True)
time2
```

```python
df.loc[(df['Formatted Date']<time2) & (df['Formatted Date']>time1)]
```

```python
df_jan = df.loc[(df['Formatted Date']<time2) & 
                (df['Formatted Date']>=time1)]
df_jan.plot(x='Formatted Date', y='Temperature (C)', 
            kind='line',figsize=(8,4), title='Temperature (C)')
```

```python
df['date'] = [x.date() for x in df['Formatted Date']]
df['month'] = [x.month for x in df['Formatted Date']]
df['year'] = [x.year for x in df['Formatted Date']]
df['day'] = [x.day for x in df['Formatted Date']]
df['hour'] = [x.hour for x in df['Formatted Date']]
```

```python
df
```

```python
df.groupby('year').sum('Temperature (C)')
```

```python
df['day_cos'] = [np.cos(x * (2 * np.pi / 24)) for x in df['hour']]
df['day_sin'] = [np.sin(x * (2 * np.pi / 24)) for x in df['hour']]
```

```python
# Seconds in day 
s = 24 * 60 * 60 
# Seconds in year 
year = (365.25) * s

df['year_cos'] = [np.cos((x) * (2 * np.pi / year)) for x in df['timestamp']]
df['year_sin'] = [np.sin((x) * (2 * np.pi / year)) for x in df['timestamp']]
```
