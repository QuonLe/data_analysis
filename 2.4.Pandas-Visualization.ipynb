```python
# confusion matrix

import pandas as pd

data = {'Actual Value':    [1, 0, 0, 1, 0, 1, 0, 0, 1],
        'Predicted Value': [0, 1, 0, 1, 0, 1, 1, 0, 1]}
df = pd.DataFrame(data)
df
```

```python
confusion_matrix = pd.crosstab(df['Predicted Value'], df['Actual Value'],
                               rownames=['Predicted'], colnames=['Actual'])
print(confusion_matrix)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib qt

sn.heatmap(confusion_matrix, annot=True)
plt.show()
```
